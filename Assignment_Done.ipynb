### 6. Find the total sales and total profit for each region

```python
# Changing sales and profit columns data types to integer data types   

# Remove leading and trailing spaces from the 'Sales' column
df['Sales'] = df['Sales'].str.strip()

# Convert the 'Sales' column to numeric type
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')

# Convert the 'Profit' column to numeric type
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
```

```python
# Total sales by region

total_sales = df.groupby('Region')['Sales'].sum()
print("Here is the total sales by region:\n", total_sales)
```

```python
# Total profit by region

total_profit = df.groupby('Region')['Profit'].sum()
print("Here is the total profit by region:\n", total_profit)
```

### 7. Plot a Bar Chart for the regional distribution of sales and profit

```python
# grouping the sales by region
region_sales = df.groupby('Region').sum().reset_index()
```

```python
# Plot a bar chart for regional distribution of sales
plt.figure(figsize=(8, 6))
plt.bar(region_sales['Region'], region_sales['Sales'], color='purple')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Regional Distribution of Sales')

plt.show()
```

```python
# grouping the Profit by region
region_profit = df.groupby('Region').sum().reset_index()
```

```python
# Plot a bar chart for regional distribution of Profit
plt.figure(figsize=(8, 6))
plt.bar(region_profit['Region'], region_profit['Profit'], color='skyblue')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.title('Regional Distribution of Profit')

plt.show()
```

### 12. Find the total sales and total profit for each year

```python
#converting order date and ship date to proper date format
df['Order Date'] = pd.to_datetime(df['Order Date'])


df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df.info()
```

```python
#extracting the year from the ship date column
df['Ship Year'] = pd.to_datetime(df['Order Date']).dt.year
```

```python
# Total sales by by year
sales_by_year = df.groupby('Ship Year')['Sales'].sum().reset_index()
print('The total sales by:\n', sales_by_year)
```

```python
# Total profit by by year
profit_by_year = df.groupby('Ship Year')['Profit'].sum().reset_index()
print('The total profit by:\n', profit_by_year)
```

13. ### Plot a line Chart for the yearly distribution of sales and profit

```python
# plotting a line chart for the yearly distribution of sales

plt.plot(sales_by_year['Ship Year'],sales_by_year['Sales'], color="blue", marker = 'o')

# Adding data labels
for i, revenue in enumerate(sales_by_year['Sales']):
    plt.text(sales_by_year['Ship Year'][i], revenue, f'{revenue}', fontsize=10, color='black', ha='center')
    
plt.title('Total Sales per Ship Year',fontdict={'family': 'sans-serif', 'size': 16, 'color': 'red'})
plt.xlabel('Ship Year', fontdict={'family': 'sans-serif', 'size': 14, 'color': 'red'})
plt.ylabel('Sales', fontdict={'family': 'sans-serif', 'size': 14, 'color': 'red'})
plt.show()
```

```python
# plotting a line chart for the yearly distribution of profit

plt.plot(profit_by_year['Ship Year'],profit_by_year['Profit'], color="green", marker = 'o')   
plt.title('Total Profit per Ship Year',fontdict={'family': 'sans-serif', 'size': 16, 'color': 'blue'})
plt.xlabel('Ship Year', fontdict={'family': 'sans-serif', 'size': 14, 'color': 'blue'})
plt.ylabel('Profit', fontdict={'family': 'sans-serif', 'size': 14, 'color': 'blue'})
plt.show()
```
